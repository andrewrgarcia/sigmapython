```python
from frame_pdsfit import *
import numpy as np
```

The core function of frame_pdsfit is **make()**, which has the form:

def **make**(data,name,pds=['gauss','lognorm','expon','gamma','beta'], plots=True,bins= 15,xlims=[0,400])

**make()** takes the following inputs:

* 1\. **data** - this is the input which holds a vector of all your histogram data <br/>
* 2\. **name** - any name you would want to give to your specific data <br/>
* 3\. **pds**  - this is an input vector which specifies which distribution types you want to fir to your data; it is defaulted to all the distribution types I have declared in the script, which are: <br/>
    * a\. gauss - Gaussian / normal distribution <br/>
    * b\. lognorm - Lognormal distribution <br/>
    * c\. expon - Exponential distribution <br/>
    . . . You get the point. Only include the names of the distributions you want to fit to your data.  <br/>
* 4\. **plots** - Boolean to plot (or not) <br/>
* 5\. **bins** - number of bins your histogram holds <br/>
* 6\. **xlims** - x-axis limits to plot. If xlims = '' computer makes best decision of x-limits 

Let's make 3 random sets of data in the following way:

```python
ex1 = np.random.normal(10, 10, 1000)
ex2 = 2*np.random.uniform(1,40, 1000) + np.random.normal(10, 10, 1000)
ex3 = 3*np.random.exponential(4,100)
```

Now let's fit these examples to different distribution functions with **make**: 

```python
make(ex1,'example 1',['gamma','beta'],bins=20,xlims='')
```

```python
make(ex2,'example 2',['gauss'],bins=20,xlims='')
```

```python
make(ex3,'example 3',bins=20,xlims='')
```
